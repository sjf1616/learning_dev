```python
import numpy as np

#Dados fictícios 
x = np.array([1, 2, 3, 4])
y = np.array([3, 5, 7, 9])

#Hiperparâmetros
alpha = 0.01
epochs = 1000

#Inicialização dos parâmetros
m = 0
b = 0
n = len(x)

#Gradiente Descendente
for _ in range(epochs):
    y_pred= m * x + b
    error = y - y_pred
    dm = (-2/n) * np.dot(x, error)
    db = (-2/n) * np.sum(error)

    m = m - alpha * dm
    b = b - alpha * db

print(f'm: {m:.4f}, b: {b:.4f}')
```

```python
import matplotlib.pyplot as plt

# Simulação (valores fictícios para o gráfico)
epochs = list(range(20))
loss_small = [3000 - i*20 for i in epochs]         # α = 0.001 (devagar)
loss_good  = [3000 / (i+1) for i in epochs]        # α = 0.01 (ideal)
loss_big   = [3000 - (i*200*(-1)**i) for i in epochs]  # α = 1.0 (instável)

plt.plot(epochs, loss_small, label='α = 0.001 (pequeno)')
plt.plot(epochs, loss_good,  label='α = 0.01  (ideal)')
plt.plot(epochs, loss_big,   label='α = 1.0   (grande)')
plt.xlabel('Epochs')
plt.ylabel('Erro (Loss)')
plt.legend()
plt.title('Comparando diferentes learning rates')
plt.grid()
plt.show()
```
